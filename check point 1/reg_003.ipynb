```python
# Outlier plotting
outlier_indexes = []

def outlier_plotting(feature):
    outlier = []
    # Plotting section
    plt.figure(figsize=(6,3))
    # Replacing d_train with train_data
    sns.boxplot(x=train_data[feature], palette="Set3")
    plt.title("{}'s Outlier Box Plot".format(feature), weight = "bold")
    plt.xlabel(feature, weight = "bold")
    plt.show()

    # Outlier computing
    Q1 = train_data[feature].quantile(0.25)
    Q3 = train_data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_outlier_limit = Q1 - (1.5 * IQR)
    higher_outlier_limit = Q3 + (1.5 * IQR)

    print("Values lower than {} and higher than {} are outliers for {}.\n".format(lower_outlier_limit,higher_outlier_limit,feature))

    # There are different ways to detect and show outlier values, I will use Z-Score method instead of writing conditional function.

    # Outlier detecting
    threshold = 3

    for i in train_data[feature]:
        z = (i-train_data[feature].mean())/train_data[feature].std()
        if z > threshold:
            outlier.append(i)
            index = train_data[train_data[feature] == i].index[0]
            outlier_indexes.append(index)
    if outlier == []:
        print("No any outliers for {}.".format(feature))
    else:
        print("There are {} outliers for {}:".format(len(outlier),feature), outlier)

for i in [col for col in train_data.columns if train_data[col].dtype != 'O']:
    if i != "Id":
        outlier_plotting(i)
```

```python
# Handling Missing Values and Feature Engineering
alldata = pd.concat([train_data,test_data],axis=0,sort=False)
alldata["SalePrice"].head()
```

```python
pd.set_option('display.max_rows', 100)
info_count = pd.DataFrame(alldata.isnull().sum(),columns=['Count of NaN'])
dtype = pd.DataFrame(alldata.dtypes,columns=['DataTypes'])
info = pd.concat([info_count,dtype],axis=1)
info
```

```python
from sklearn.preprocessing import LabelEncoder

# Filling 433 LotFrontage values. I will use linear interpolation to fill these NaN values.
alldata['LotFrontage'].interpolate(method='linear',inplace=True)

# Filling other NaNs
for i in info.T:
    if i == "Id" or i == "SalePrice" or i == "LotFrontage":
        continue
    else:
        if (info.T[i][0] == 0):
            continue
        elif (info.T[i][0] < 400):
            alldata[i].fillna(alldata[i].value_counts().index[0], inplace = True)
        else:
            lbl_enc = LabelEncoder()
            lbl_enc.fit(list(alldata[i].values))
            alldata[i] = lbl_enc.transform(list(alldata[i].values))
```

```python
# We have still columns in object datatype. We need to handle this columns. Because we will use machine learning algorithms.
list_ = ["MSZoning", "Street", "LotShape", "LandContour", "Utilities", "LotConfig",
        "LandSlope", "Neighborhood", "Condition1", "Condition2", "BldgType", "HouseStyle",
        "RoofStyle", "RoofMatl", "Exterior1st", "Exterior2nd",
        "MasVnrType", "ExterQual", "ExterCond", "Foundation", "BsmtQual", "BsmtCond", "BsmtExposure",
        "BsmtFinType1", "BsmtFinType2", "Heating", "HeatingQC", "CentralAir", "Electrical", "KitchenQual",
        "Functional", "GarageType", "GarageFinish", "GarageQual", "GarageCond", "PavedDrive", "SaleType",
        "SaleCondition"]

for feature in list_:
    alldata[feature]= alldata[feature].astype("category")
    alldata = pd.get_dummies(alldata, columns=[feature])
```

# **Modeling and Evaluation**

```python
test = test.drop("SalePrice", axis=1)
```

```python
y = np.log1p(train['SalePrice'])
x = train.drop(["Id", "SalePrice"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn import metrics
models = [('LR', LinearRegression()),
          ("Ridge", Ridge()),
          ("Lasso", Lasso()),
          ("ElasticNet", ElasticNet()),
          ('KNN', KNeighborsRegressor()),
          ('RF', RandomForestRegressor()),
          ('SVR', SVR()),
          ('GBR', GradientBoostingRegressor())]
```

```python
for name, regressor in models:
    model = regressor
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    MSE = metrics.mean_squared_error(y_test,y_pred)
    RMSE = np.sqrt(MSE)
    R2 = metrics.r2_score(y_test, y_pred)
    print(f"{name} - RMSE: {round(RMSE, 4)}, R2: {round(R2, 4)}")
```

```python
gbr = models[1][1]

X_test = test.drop(columns=['Id'])
y_pred = gbr.predict(X_test)

predictions_df = pd.DataFrame({
    'Id': test['Id'],
    'SalePrice': np.floor(np.exp(y_pred)).astype(int)
})

predictions_df.to_csv('/Users/neehanthreddym/Desktop/AI/Project/Result/REG-03-CKPT1.csv', index=False)
```

## **Summary**

We started by separate numerical and categorical features according to distinct value counts in order to provide efficient feature analysis for home price prediction. To better understand data variation and alignment across training and test sets, countplots and histograms were used to visualize categorical and numerical distributions, respectively. IQR and Z-score techniques were used to identify outliers, and box plots were included for visualization. For numerical data, missing values were handled by interpolation; for categorical values, they were filled with the most common values or encoded using get_dummies and LabelEncoder. Heatmap-based correlation analysis revealed connections with SalePrice, which helped choose features. We divided the data, normalized SalePrice using log transformation, and trained several regression models (e.g., Random Forest, Gradient Boosting, Linear Regression). Lastly, in order to determine which model was the greatest predictor of house prices, we used Root Mean Squared Error (RMSE) to evaluate the models. Then exported the predicted values into a CSV file, which was based on the best model for this data.
